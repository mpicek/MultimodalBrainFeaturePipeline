```python
from deeplabcut.utils.auxfun_videos import imread
from deeplabcut.pose_estimation_tensorflow.core.predict import extract_cnn_output, multi_pose_predict, argmax_pose_predict
import numpy as np
import sys
sys.path.append('/repo')
from data_synchronization.dlc_utils import prepare_network

cfg, dlc_cfg, sess, inputs, outputs, pdindex, DLCscorer, DLCscorerlegacy, start_path = prepare_network(
    config=config_path,
    directory='/data/UP2_LED_detector-mpicek-2024-03-20/labeled-data/output/',
    frametype=".png",
    shuffle=1,
    trainingsetindex=0,
    gputouse=1,
    save_as_csv=True,
    modelprefix="",
)

def getpose(image, cfg, sess, inputs, outputs, outall=False):
    """Extract pose"""
    im = np.expand_dims(image, axis=0).astype(float)
    outputs_np = sess.run(outputs, feed_dict={inputs: im})
    scmap, locref = extract_cnn_output(outputs_np, cfg)
    num_outputs = cfg.get("num_outputs", 1)
    if num_outputs > 1:
        pose = multi_pose_predict(scmap, locref, cfg["stride"], num_outputs)
    else:
        pose = argmax_pose_predict(scmap, locref, cfg["stride"])
    if outall:
        return scmap, locref, pose
    else:
        return pose
```

```python
img_path = '/data/UP2_LED_detector-mpicek-2024-03-20/labeled-data/output/img496.png'
image = imread(img_path, mode="skimage")
prediction = getpose(image, dlc_cfg, sess, inputs, outputs)
```

```python
forehead_x, forehead_y, forehead_confidence = prediction[0][0], prediction[0][1], prediction[0][2]
nose_x, nose_y, nose_confidence = prediction[1][0], prediction[1][1], prediction[1][2]
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import cv2

# Load the image
image = cv2.imread(img_path)

# # Draw a red point at (x, y) with a thickness of 2 pixels
color_forehead = (0, 0, 255)  # Red color in BGR format
color_nose = (0, 255, 255)  # Red color in BGR format
thickness = 2
cv2.circle(image, (int(forehead_x), int(forehead_y)), thickness, color_forehead, 3)
cv2.circle(image, (int(nose_x), int(nose_y)), thickness, color_nose, 3)

cv2.imshow('image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
# plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
