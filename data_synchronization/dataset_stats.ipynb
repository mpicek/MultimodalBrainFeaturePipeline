```python
import numpy as np
import os
kinematics_folder = '/media/vita-w11/T73/martin/data_final/kinematics'
datapoint_len = 30

overall_length = 0

for file in os.listdir(kinematics_folder):
    if file.endswith('_in_dataset.npy'):
        in_dataset = np.load(os.path.join(kinematics_folder, file))

        convolved = np.convolve(in_dataset, np.ones(datapoint_len), mode='valid')
        possible_indices = np.where(convolved == datapoint_len)[0]

        last_index = -np.inf # so that the first index is included
        dataset_indices = []
        for index in possible_indices:
            if index > last_index + datapoint_len:
                dataset_indices.append(index)
                last_index = index
        overall_length += len(dataset_indices)
        print(f'{file}: {len(dataset_indices)}, {round(100 * len(dataset_indices) / (len(in_dataset) / datapoint_len), 2)}%')
```
